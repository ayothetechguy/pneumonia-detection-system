```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import os
from pathlib import Path
import random

# Set style
plt.style.use('default')
sns.set_palette("husl")

# Display settings
%matplotlib inline
plt.rcParams['figure.figsize'] = (15, 10)

print("✅ Libraries imported successfully!")
```

```python
# Define paths
BASE_DIR = Path('/Users/ayoolumimelehon/Desktop/chest-xray-pneumonia/data/chest_xray')
TRAIN_DIR = BASE_DIR / 'train'
TEST_DIR = BASE_DIR / 'test'
VAL_DIR = BASE_DIR / 'val'

# Check if paths exist
print(f"Base directory exists: {BASE_DIR.exists()}")
print(f"Train directory exists: {TRAIN_DIR.exists()}")
print(f"Test directory exists: {TEST_DIR.exists()}")
print(f"Val directory exists: {VAL_DIR.exists()}")
```

```python
# Count images in each set
def count_images(directory):
    normal_count = len(list((directory / 'NORMAL').glob('*.jpeg')))
    pneumonia_count = len(list((directory / 'PNEUMONIA').glob('*.jpeg')))
    return normal_count, pneumonia_count

train_normal, train_pneumonia = count_images(TRAIN_DIR)
test_normal, test_pneumonia = count_images(TEST_DIR)
val_normal, val_pneumonia = count_images(VAL_DIR)

# Create summary dataframe
data_summary = pd.DataFrame({
    'Set': ['Training', 'Testing', 'Validation'],
    'Normal': [train_normal, test_normal, val_normal],
    'Pneumonia': [train_pneumonia, test_pneumonia, val_pneumonia]
})

data_summary['Total'] = data_summary['Normal'] + data_summary['Pneumonia']
data_summary['Normal %'] = (data_summary['Normal'] / data_summary['Total'] * 100).round(2)
data_summary['Pneumonia %'] = (data_summary['Pneumonia'] / data_summary['Total'] * 100).round(2)

print("📊 Dataset Summary:")
print(data_summary)
print(f"\n🎯 Total Images: {data_summary['Total'].sum()}")
```

```python
# Create visualization of class distribution
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Bar plot
data_summary_melted = data_summary.melt(id_vars='Set', value_vars=['Normal', 'Pneumonia'], 
                                         var_name='Class', value_name='Count')
sns.barplot(data=data_summary_melted, x='Set', y='Count', hue='Class', ax=axes[0])
axes[0].set_title('Image Distribution Across Sets', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Number of Images')

# Pie chart for training set
train_data = [train_normal, train_pneumonia]
axes[1].pie(train_data, labels=['Normal', 'Pneumonia'], autopct='%1.1f%%', 
            startangle=90, colors=['#2ecc71', '#e74c3c'])
axes[1].set_title('Training Set Class Distribution', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\n⚠️ Class Imbalance Ratio: {train_pneumonia/train_normal:.2f}:1 (Pneumonia:Normal)")
```

```python
# Function to display sample images
def display_sample_images(directory, class_name, num_samples=5):
    """Display random sample images from a class"""
    class_dir = directory / class_name
    image_files = list(class_dir.glob('*.jpeg'))
    
    if len(image_files) == 0:
        print(f"No images found in {class_dir}")
        return
    
    # Randomly sample images
    samples = random.sample(image_files, min(num_samples, len(image_files)))
    
    fig, axes = plt.subplots(1, num_samples, figsize=(20, 4))
    fig.suptitle(f'{class_name} X-Ray Samples', fontsize=16, fontweight='bold', y=1.02)
    
    for idx, img_path in enumerate(samples):
        img = Image.open(img_path)
        axes[idx].imshow(img, cmap='gray')
        axes[idx].axis('off')
        axes[idx].set_title(f'{img_path.name}\n{img.size[0]}x{img.size[1]}', fontsize=10)
    
    plt.tight_layout()
    plt.show()

# Display Normal X-rays
print("🫁 NORMAL Chest X-Rays:")
display_sample_images(TRAIN_DIR, 'NORMAL', num_samples=5)

print("\n🦠 PNEUMONIA Chest X-Rays:")
display_sample_images(TRAIN_DIR, 'PNEUMONIA', num_samples=5)
```

```python
# Analyze image dimensions and properties
def analyze_image_properties(directory):
    """Analyze image dimensions across all images"""
    widths = []
    heights = []
    
    for class_name in ['NORMAL', 'PNEUMONIA']:
        class_dir = directory / class_name
        for img_path in class_dir.glob('*.jpeg'):
            img = Image.open(img_path)
            widths.append(img.size[0])
            heights.append(img.size[1])
    
    return widths, heights

print("🔍 Analyzing image properties...")
train_widths, train_heights = analyze_image_properties(TRAIN_DIR)

# Create dataframe
img_properties = pd.DataFrame({
    'Width': train_widths,
    'Height': train_heights
})

print("\n📏 Image Dimension Statistics:")
print(img_properties.describe())

# Plot distributions
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

axes[0].hist(train_widths, bins=50, color='skyblue', edgecolor='black')
axes[0].set_title('Width Distribution', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Width (pixels)')
axes[0].set_ylabel('Frequency')
axes[0].axvline(np.mean(train_widths), color='red', linestyle='--', label=f'Mean: {np.mean(train_widths):.0f}')
axes[0].legend()

axes[1].hist(train_heights, bins=50, color='lightcoral', edgecolor='black')
axes[1].set_title('Height Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Height (pixels)')
axes[1].set_ylabel('Frequency')
axes[1].axvline(np.mean(train_heights), color='red', linestyle='--', label=f'Mean: {np.mean(train_heights):.0f}')
axes[1].legend()

plt.tight_layout()
plt.show()

print(f"\n✅ Most images are around {np.mean(train_widths):.0f} x {np.mean(train_heights):.0f} pixels")
```
